```python
## A class representing the different piece on the board (red, green, blue, block

class Piece:
    # initialise the piece with type and position on the board
    def __init__(self,type_, q, r):
        self.type = type_
        self.q = q
        self.r = r
        
    # return the type of the piece
    def getType(self): return self.type
    
    #return the Q postion on the board of the piece
    def getQ(self): return self.q
    
    #return the R postion on the board of the piece
    def getR(self): return self.r
    
    #set the type of piece
    def setType(self, newType):
        self.type = newType
        
    #set the q position of the piece
    def setQ(self, newQ):
        self.q = newQ
    #set the r position   
    def setR(self, newR):
        self.r = newR
    
    #move the piece to a new position
    def moveTo(self, newQ, newR):
        self.q = newQ
        self.r = newR
    
    
```

```python
class Cell:
    
    def __init__(self, q, r):
        self.q = q
        self.r = r
        self.node = None
        self.occupied = False
    
    def addPiece(self, piece):
        self.node = piece
        self.occupied = True
        
    def removePiece(self):
        self.node = None
        self.occupied = False
        
    def isOccupied(self):
        return self.occupied
    
```

```python
# Node for path finding in the map
class Node:
    def __init__(self, q, r):
        self.q = q
        self.r = r
        self.visited = False
        self.last = (q, r)
        
    def isVisited(self):
        return self.visited
    
    def visit(self, lastq, lastr):
        self.visited = True
        self.last = (lastq, lastr)
        print("Visited from "+ str((lastq, lastr)) + " to "+ str((self.q, self.r)))
        
    def getLastNode(self):
        return self.last
```

```python
# using heapq library to handle the priority Quee
# Heaps are binary trees for which every parent node has a value less than 
# or equal to any of its children
import heapq

class priorityQueue:
    def __init__(self):
        self.queue =[]
        self.index = 0
        
    def insert(self, priority, node):
        heapq.heappush(self.queue, (priority, self.index, node))
        self.index += 1
        
    def remove(self):
        return heapq.heappop(self.queue)[-1]
    
    def isEmpty(self):
        return (self.queue) == 0
```

```python
def checkReachGoal(endNodes, goals):
    for node in endNodes:
        for goal in goals:
            if node.q == goal.q and node.r == goal.r:
                return (True, (goal.q, goal.r))
        
    return (False,(0,0))
```

```python
def checkForExit(piece, goals):
    for goal in goals:
        if piece.getQ() == goal.q and piece.getR() == goal.r:
            return True
    return False
```

```python
def jumpActionsDetection(path):
    totaljump = 0
    for i in range(len(path) -1):
        if abs(path[i][0]  - path[i+1][0]) == 2 or abs(path[i][1] - path[i+1][1]) == 2:
            totaljump+=1
    return totaljump
```

```python
def MapGreedy(pieces, board, goals):
    exitNumber = len(pieces)
    
    while exitNumber != 0:
        exitmove = False
        
        #check for exist move:     
        for i in range(len(pieces)):
            if checkForExit(pieces[i], goals):
                piece = pieces[i]
                print("EXIT from "+str((piece.getQ(), piece.getR())))
                #remove it from the board
                board[piece.getQ()+3][piece.getR()+3].removePiece()
                del pieces[i]
                exitmove=True
                exitNumber -= 1
                # can only move one at each round
                break
        #can only move one at each round
        if exitmove:
            continue
        
        
        #for each piece find the best path to the goal
        paths = []
        for piece in pieces:
            paths.append(singleGreedy(piece, Board, goals)[::-1])
        
        #search for the path with highest number of jump
        jumps = []
        for path in paths:
            jumps.append(jumpActionsDetection(path))
        maxjump = max(jumps)
        #print(maxjump)
        
        #check for the shorttest path in the maxjump
        shortestloc = 0
        shortestlen = 999
        for i in range(len(jumps)):
            if jumps[i] == maxjump:
                if len(paths[i]) < shortestlen:
                    shortestlen = len(paths[i])
                    shortestloc = i
                    #print(i)
        #do the move!
        currentloc = paths[shortestloc][0]
        nextloc = paths[shortestloc][1]
        
        #deregister from the current node
        board[currentloc[0]+3][currentloc[1]+3].removePiece()
        
        #move the piece and register it at the new node
        pieces[shortestloc].setQ(nextloc[0])
        pieces[shortestloc].setR(nextloc[1])
        board[nextloc[0]+3][nextloc[1]+3].addPiece(pieces[shortestloc])
        
        if abs(currentloc[0] - nextloc[0]) == 2 or abs(currentloc[1] - nextloc[1]) ==2:
            print("JUMP from "+str(currentloc) + " to "+ str(nextloc))
        else:
            print("Move from "+str(currentloc) + " to "+ str(nextloc))
        
                
        
```
